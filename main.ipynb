```python
# Materials
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
SAVE_IMAGES_IN_FOLDER = True
```

## Figure 1 {-}

```python
F1 = nx.MultiDiGraph()

edgesF1 = np.array(['ab', 'bc', 'ca', 'cd'])
nodesF1 = np.array(['a', 'b', 'c', 'd'])

F1.add_nodes_from(nodesF1)
F1.add_edges_from(edgesF1)
posF1 = nx.circular_layout(F1)
nx.draw(
    F1, posF1, edge_color='gray', width=1, linewidths=1, node_size=700, node_color='black', font_color='white', alpha=1,
    labels={node: node for node in nodesF1}, label="Node", arrowstyle='-', connectionstyle='arc3,rad=0.2'
)

legend_elementsF1 = [Line2D([0], [0], color='gray', lw=1.4, label='Undirected simple payment'),
                   Line2D([0], [0], marker='o', color='w', label='Firm',
                          markerfacecolor='black', markersize=25)]

plt.legend(handles=legend_elementsF1, loc='lower right')
plt.axis('off')
plt.title("Market G")
# WARNING! THIS WILL SAVE THE IMAGE OF THE GRAPH IN A FOLDER "figures" inside this folder
SAVE_IMAGES_IN_FOLDER and plt.savefig('./figures/F1.pdf')
plt.show()
```

## Figure 2 {-}

```python
F2 = nx.MultiDiGraph()

edgesF2 = np.array(['ab', 'ba', 'bc', 'cb', 'cd', 'dc', 'da', 'ad'])
nodesF2 = np.array(['a', 'b', 'c', 'd'])

F2.add_nodes_from(nodesF2)
F2.add_edges_from(edgesF2)
posF2 = nx.circular_layout(F2)
nx.draw(
    F2, posF2, edge_color='gray', width=1, linewidths=1, node_color='black', font_color='white', alpha=1,
    labels={node: node for node in nodesF2}, label="Node", arrowstyle='->', connectionstyle='arc3,rad=0.2'
)

legend_elementsF2 = [Line2D([0], [0], color='gray', lw=1.4, label='Undirected general payment'),
                     Line2D([0], [0], marker='o', color='w', label='Firm',
                            markerfacecolor='black', markersize=19)]

plt.legend(handles=legend_elementsF2, loc='lower right')
plt.axis('off')
plt.title("Market G")
SAVE_IMAGES_IN_FOLDER and plt.savefig('./figures/F2.pdf')
plt.show()
```
