```python
# Materials
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

## Figure 1 {-}

```python
G1 = nx.MultiDiGraph()

edgesG1 = np.array(['ab', 'bc', 'ca', 'cd'])
nodesG1 = np.array(['a', 'b', 'c', 'd'])

G1.add_nodes_from(nodesG1)
G1.add_edges_from(edgesG1)
posG1 = nx.circular_layout(G1)
nx.draw(
    G1, posG1, edge_color='gray', width=1, linewidths=1, node_color='black', font_color='white', alpha=1,
    labels={node: node for node in nodesG1}, label="Node", arrowstyle='-', connectionstyle='arc3,rad=0.2'
)

legend_elementsG1 = [Line2D([0], [0], color='gray', lw=1.4, label='Undirected simple payment'),
                   Line2D([0], [0], marker='o', color='w', label='Firm',
                          markerfacecolor='black', markersize=19)]

plt.legend(handles=legend_elementsG1, loc='lower right')
plt.axis('off')
plt.title("Market G")
plt.show()
```
